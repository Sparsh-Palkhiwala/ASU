```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('tvmarketing.csv')
plt.scatter(data.studytime,data.score)
plt.show()
```

```python
def loss_function(m,b,points):
    total_error = 0
    for i in range(len(points)):
        x = points.iloc[i].studytime
        y = points.iloc[i].score
        total_error = total_error + (m*x + b) ** 2  #Recursion of the error
    total_error/float(len(points))
```

```python
def gradient_descent(m_now,b_now,points,lr):
    m_grad = 0
    b_grad = 0
    n = len(points)
    for i in range(n):
        x = points.iloc[i].studytime
        y = points.iloc[i].score
        m_grad = m_grad + (-2/n) * x * (y - (m_now*x + b_now))
        b_grad = b_grad + (-2/n) * (y-(m_now*x + b_now))
    m_new = m_now - (lr*m_grad)
    b_new = b_now - (lr*b_grad)
    return m_new,b_new

m = 0
b = 0
lr = 0.00001
epochs = 1000

for i in range(epochs):
    m,b = gradient_descent(m,b,data,lr)

print(m,b)

plt.scatter(data.studytime,data.score,color = "black")
plt.plot(list(range(0,300)), [m*x + b for x in range(0,300)], color = "red" )
plt.show()
```
